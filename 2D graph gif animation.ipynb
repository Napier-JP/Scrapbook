```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
```

```python
%matplotlib nbagg

x = np.arange(0, 10, 0.1)

def example_function(i):
    plt.clf()# 現在描画されているグラフを消去
    y = 2 * np.sin(x - i)
    plt.plot(x, y)

fig = plt.figure()
anim = FuncAnimation(fig, example_function, frames = range(0,31), interval = 200, repeat=True, blit=False)
#anim.save("sinecurve.gif", writer=PillowWriter(fps=5))
```

```python
%matplotlib nbagg

x = np.arange(0, 2 ,0.025)

def heartbeat(i):
    plt.clf()# 現在描画されているグラフを消去
    y = np.sin(x + i*0.1)**63 * np.sin(x + i*0.1 + 1.5) * 8
    plt.plot(x, y, color='green', linewidth=1)
    plt.axis('off')
    plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
    plt.plot(x, [1.2 for _ in range(80)], color='gray', linewidth=1)
    plt.plot(x, [-0.6 for _ in range(80)], color='gray', linewidth=1)

fig = plt.figure()
anim = FuncAnimation(fig, heartbeat, frames = range(-8,21), interval = 33, repeat=True, blit=False)
#anim.save("heartbeat.gif", writer=PillowWriter(fps=30))
```
